# 3 Machine Learning Pipeline

**3.1 Logistic Regression Model with 5-Fold CV Hyperparameter Tuning**

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as  sns
import plotly.express as px
import warnings

from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV, train_test_split, KFold, cross_validate
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer, KNNImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn import set_config
from sklearn.naive_bayes import CategoricalNB
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import RocCurveDisplay, roc_curve, accuracy_score, confusion_matrix, roc_auc_score, classification_report
```

```python
# Define pipelines for numerical and categorical features

# Numerical pipeline
#   1.  Implement MICE imputation for missing values
#   2.  Normalize values using MinMaxScaler
numerical_pipeline = Pipeline(steps = [
    ('imputer', IterativeImputer(max_iter = 10, random_state = 42)),
    ('scaler', MinMaxScaler())
])

# Categorical pipeline
#   1.  Implement K-NN imputation for missing values
#   2.  One-hot encode categorical features
categorical_pipeline = Pipeline(steps = [
    ('imupter', KNNImputer(n_neighbors = 3)),
    ('one-hot', OneHotEncoder(handle_unknown = 'ignore', sparse = False))
])
```

```python
# Define pipeline to transform numerical and categorical feature columns with respective pipelines
column_transform = ColumnTransformer(transformers = [
    ('numerical_pipeline', numerical_pipeline, numerical_columns),
    ('categorical_pipeline', categorical_pipeline, categorical_columns)
],
                                     remainder = 'drop',
                                     n_jobs = -1
                                     )
```

```python
# Define Logistic Regression model pipeline
model = LogisticRegression(max_iter = 1000, random_state = 42)
model_pipeline = Pipeline(steps = [
    ('column_transform', column_transform),
    ('model', model)
])
```

```python
# Display Logistic Regression model pipeline
set_config(display = 'diagram')
display(model_pipeline)
```

```python
# Prepare data
data_x = data[numerical_columns + categorical_columns]
data_y = data[target_column]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=42)

# Cross-Validation using K-Folds (k = 5)
k = 5

# Define hyperparameter tuning for Logistic Regression model
param_grid = {'model__penalty' : ['l1', 'l2'],
              'model__C' : np.logspace(-4, 4, 20),
              'model__solver' : ['liblinear']}

# Initiate grid search
cv = GridSearchCV(model_pipeline, param_grid, cv = k, n_jobs = -1)

# Fit to training set
cv.fit(X_train, y_train)

# Predict the labels of the test set to y_pred
y_pred = cv.predict(X_test)

# Plotting the confusion matrix for the tuned model against testing data
classes = data['Diabetes'].unique()
sns.heatmap(confusion_matrix(y_test, y_pred, labels = classes), annot = True,
            xticklabels = classes, yticklabels = classes, cmap = 'Reds',
            fmt = 'g')

plt.xlabel('Predicted')
plt.ylabel('Actual')
title = 'Confusion Matrix\npen = l1; C = 0.616; solver = liblinear;\ndata_bal_met = none'
plt.title(title)

plt.show()

# Compute and print metrics
print("Accuracy: {}".format(cv.score(X_test, y_test)))
print(classification_report(y_test, y_pred))
print("Tuned Model Parameters: {}".format(cv.best_params_))
```

```python
# In-depth results per test split in 5-fold CV grid search
results = pd.DataFrame(cv.cv_results_)
results[results['rank_test_score'] == 1]
```

```python
# Display intecept and regression coefficients of the tuned model
print('The intercept is: %.5f' % cv.best_estimator_.named_steps.model.intercept_)
print('The regression coefficients are: ' + str(cv.best_estimator_.named_steps.model.coef_))
```
